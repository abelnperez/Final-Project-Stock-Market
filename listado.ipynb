```python
df['Code'] = df['Ticker'].apply(lambda x: re.search(r'\((.*?)\)', x).group(1) if re.search(r'\((.*?)\)', x) else None)
```

```python
df
```

```python
# Reordenamos las columnas para que "Code" sea la primera columna
df = df[['Code'] + [col for col in df.columns if col != 'Code']]

df
```

```python
# Eliminar duplicados basados en la columna 'codigo'
df_sin_duplicados = df.drop_duplicates(subset='Code', keep='first')
```
